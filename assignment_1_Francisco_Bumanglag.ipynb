**Q4**: Convert the Columns with "Yes"/"No" entries into Boolean columns. Use a for loop over the names of the relevant columns and the method .map().

```python
#iterates through each column in the DataFrame smh. For each column, 
#it checks if the unique values in that column are exactly equal to the set {'Yes', 'No'}. 
#If they are, it maps 'Yes' to True and 'No' to False within that column, effectively converting it to Boolean values. 
#If not, it skips the column without making any changes.

#https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.unique.html
#https://pandas.pydata.org/docs/reference/api/pandas.Series.map.html

#for loop to iterate through all columns in the DataFrame
for column in smh.columns:
    if set(smh[column].unique()) == {'Yes', 'No'}:  #retrieves all unique values from the columns. smh.set(...): converts the unique values into a set
        smh[column] = smh[column].map({'Yes': True, 'No': False})
        
#display the updated DataFrame
smh.head()
```

        For example: "3.50 - 4.00" should have a value of 3.75 instead. Notice that the ranges in the "GPA" column are currently strings.

```python
#To calculate the average between endpoints in a specific column of a DataFrame, 
#you would need to implement custom logic to parse the string representations of the ranges and compute the averages. 
#This involves splitting the string into endpoints and calculating the average.

#1. Iterate through the DataFrame column containing the string ranges
#2. For each row, split the string into two endpoints
#3. Convert the endpoints to numerical values (e.g., float)
#4. Calculate the average of the endpoints
#5. Replace the original string with the calculated average

import pandas as pd  # Don't forget to import Pandas

#sample DataFrame with a 'GPA' column containing string ranges
data = {'GPA': ['2.00 - 2.99', '3.00 - 3.49', '3.50 - 4.00']}
df = pd.DataFrame(data)

#function to calculate the average between endpoints
def calculate_average(range_str):
    endpoints = range_str.split('-')         #splits it into a list of substrings using '-' as the delimiter
    endpoint1 = float(endpoints[0].strip())  #remove any leading or trailing whitespace from each endpoint and convert them to floats
    endpoint2 = float(endpoints[1].strip())  #remove any leading or trailing whitespace from each endpoint and convert them to floats
    average = (endpoint1 + endpoint2) / 2.0  #calculate the average of the two endpoints
    return average

#apply the function to the 'GPA' column


df['GPA'] = df['GPA'].apply(calculate_average)

print(df)

```

```python
#1. Create a sample DataFrame with 'GPA' and 'N_disorder' columns.
#2. Calculate the correlation coefficient between 'GPA' and 'N_disorder'.
#3. Store the correlation coefficient in 'correlation_coefficient'.
#4. Print the correlation coefficient between 'GPA' and 'N_disorder'.


#sample DataFrame with 'GPA' and 'N_disorder' columns -- numeric fields 
data = {'GPA': [3.00, 3.50, 2.75, 3.25, 2.90],
        'N_disorder': [0, 1, 2, 1, 3]}

df = pd.DataFrame(data)   

#compute the correlation coefficient
#np.corrcoef(df['GPA'], df['N_disorder']) computes the correlation matrix between the 'GPA' and 'N_disorder' columns
#[0, 1] is used to access the correlation coefficient value at the intersection of the first row (index 0)
#and the second column (index 1) in the correlation matrix, which represents the correlation between 'GPA' and 'N_disorder.'
corr_coef = np.corrcoef(df['GPA'], df['N_disorder'])[0, 1]

#display the result
print("Correlation Coefficient between GPA and N_disorder:", corr_coef)   
```
